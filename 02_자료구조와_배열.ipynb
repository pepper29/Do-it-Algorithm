```python
# 실습 2-2

from typing import Any, Sequence

def max_of(a: Sequence) -> Any:
    #시퀀스형 a원소의 최대값을 반환
    maximum = a[0]
    for i in range(1, len(a)):
        if a[i] > maximum:
            maximim = a[i]
    return maximim

if __name__== '__main__':
    print('배열의 최대값을 구한다')
    num = int(input('원소 수를 입력: '))
    x = [None] * num # 원소 수가 num인 리스트를 생성

    for i in range(num):
        x[i] = int(input(f'x[{i}]값은? '))

    print(f'최대값은 {max_of(x)}')
```

```python
# 2-3
# 배열 원소의 최대값을 구해서 출력


def max_of(a: Sequence) -> Any:
    #시퀀스형 a원소의 최대값을 반환
    maximum = a[0]
    for i in range(1, len(a)):
        if a[i] > maximum:
            maximim = a[i]
    return maximim


print('배열의 최대값')
print('end 입력시 종료')

number = 0
x = []  # 빈리스트

while True:
    s = input(f'x[{number}]값은? ')
    if s == 'end':
      break
    x.append(int(s))
    number += 1

print(f'{number}개 입력')
print(f'최대값은{max_of(x)}')
```

```python
# 실습 2-4

# 배열 원소의 최대값 구해서 출력

import random

def max_of(a: Sequence) -> Any:
    #시퀀스형 a원소의 최대값을 반환
    maximum = a[0]
    for i in range(1, len(a)):
        if a[i] > maximum:
            maximim = a[i]
    return maximim

print('난수의 최대값은?')
num = int(input('난수개수: '))
lo = int(input('난수 최소값: '))
hi = int(input('난수 최대값: '))
x = [None] * num  # 원소 수가 num인 리스트를 생성

for i in range(num):
  x[i] = random.randint(lo, hi) #lo 이상 hi 이하인 정수 난수를 반환

print(f'{(x)}')
print(f'최대값은 {max_of(x)}')
```

```python
# 기수와 서수
# 기수는 수를 나타내는 기초이며, 10진법에서는 0-9까지의 정수이다.
# 서수는 사물의 순서를 나타낸다.
```

```python
# 실습 2-7 A
# 10진수 정수값을 입력받아 2-36진수로 변환하여 출력
def card_conv(x: int, r: int) -> str:
    """정수값 x를 r진수로 변환하고 문자열 반환"""

    d = '' # 변환 후의 문자열
    dchar = '0123456789ABCDEFGHIJKLMNOPQRSTUVWXYZ'

    while x > 0:
        d += dchar[x % r] # 해당하는 문자를 꺼내 결합
        x //= r

    return d[::-1]  # 역순으로 반환
```

```python
# 실습 2-7 B
if __name__ == '__main__':
    print('10진수를 n진수로 변환')

    while True:
      while True: #음이 아닌 정수를 입력받음
          no  = int(input('변환할 값으로 음이 아닌 정수를 입력하세요.'))
          if no > 0:
            break

      while True:
          cd = int(input('어떤 진수로 변환?'))
          if 2 <= cd <= 36:
            break

      print(f'{cd}진수로는 {card_conv(no, cd)}')

      retry = input('한번 더 변환? y/n')
      if retry in {'n', 'N'}:
        break
```

```python
# 소수 나열하기
# 실습 2-8

counter = 0 # 나눗셈 횟수를 카운트

for n in range(2, 1001):
    for i in range(2, n):
        counter += 1
        if n % i == 0: # 나누어 떨어지면 소수가 x
            break # 반복은 더 이상 불필요
    else:
        print(n)
print(f'나눗셈 횟수: {counter}')
```

```python
# 1000 이하의 소수를 나열

counter = 0   # 나눗셈 횟수를 카운트
ptr = 0       # 이미 찾은 소수의 개수
prime = [None] * 500 # 소수를 저장하는 배열


prime[ptr] = 2  # 2는 소수이므로 초깃값으로 지정
ptr += 1

for n in range(3, 1001, 2):   # 홀수만을 대상으로 설정
    for i in range(1, ptr):   # 이미 찾은 소수로 나눔
        counter += 1
        if n % prime[i] == 0:  # 나눠지면 소수x
            break

    else:
        prime[ptr] = n # 소수로 배열에 등록
        ptr += 1

for i in range(ptr):    # ptr의 소수를 출력
    print(prime[i])
print(f'나눗셈을 실행한 횟수: {counter}')
```

```python
# 실습 2-10

# 1000이하의 소수를 나열하기

counter = 0   # 곱셈과 나눗셈을 합한 횟수
ptr = 0       # 이미 찾은 소수 개수
prime = [None] * 500 # 소수를 저장하는 배열


prime[ptr] = 2    # 시작 소수
ptr += 1

prime[ptr] = 3    # 3도 소수
ptr += 1

for n in range(5, 1001, 2): # 홀수만이 대상
    i = 1
    while prime[i] * prime[i] <= n:
        counter += 2
        if n % prime[i] ==0:  # 나누어떨어지므로 소수가 아닐 때
            break
        i += 1
    else:
        prime[ptr] = n
        ptr += 1
        counter += 1

for i in range(ptr):
    print(prime[i])
print(f'곱셈, 나눗셈을 실행횟수: {counter}')
```
